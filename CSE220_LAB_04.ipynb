```python
class Node:
  def __init__(self, e, n, p):
    self.element = e
    self.next = n
    self.prev = p
```

```python
class DoublyList:
  def __init__(self, a):
    self.arr=a
    self.head=Node(None, None, None)
    tail= self.head
    for i in range(len(a)):
      nN= Node(a[i], None, None)
      tail.next=nN
      nN.prev=tail
      tail=tail.next
    tail.next=self.head
    self.head.prev=tail
  def countNode(self):
    start= self.head.next
    counter=0
    while start!=self.head:
      counter+=1
      start=start.next
    return counter
  def forwardprint(self):
    start= self.head.next
    while start!= self.head:
      if(start.next==self.head):
        print(start.element)
      else:
        print(start.element,end=",")
      start=start.next
  def backwardprint(self):
    start= self.head.prev
    while start!=self.head:
      if(start.prev==self.head):
         print(start.element)
      else:
        print(start.element,end=",")
      start=start.prev
  def nodeAt(self, idx):
    start=self.head
    if(idx<0 or idx>=len(self.arr)):
      return None
    else:
      counter=0
      start=self.head.next
      while start!=self.head:
        if(counter==idx):
          return start
        else:
          start=start.next
          counter+=1
  def indexOf(self, elem):
    start=self.head.next
    counter=0
    result=False
    while start!=self.head:
      if(elem== start.element):
        return counter
      else:
        counter+=1
        start=start.next
    if(result==False):
      return -1 
  def insert(self, elem, idx):
    start=self.head.next
    if(idx<0 or idx>self.countNode()):
      return None
    else:
      for i in range(0,idx+1):
        if(i==idx):
          new=start
          previous=start.prev
          start=Node(elem,new,previous)
          new.prev=start
          previous.next=start
        else:
          start=start.next
  def remove(self, idx):
    if(idx<0 or idx>self.countNode()):
      return None
    else:
      start=self.nodeAt(idx)
      start.prev.next=start.next
      start.next.prev=start.prev
      return f"{start.element}"
```
